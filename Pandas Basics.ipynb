# **Data visualisation**

```python
import matplotlib.pyplot as plt
```

**Multiple charts**

```python
fig=plt.figure(figsize=(14,7))
plt.subplot(1,3,1)
plt.plot(data['SL'],data['SW'],color='green')
plt.xlabel('SL')
plt.ylabel('SW')
plt.title('Line chart')


plt.subplot(1,3,2)
plt.bar(data['PW'],data['PL'],width=0.08)
plt.grid(True)
plt.title('bar chart')


plt.subplot(1,3,3)
plt.scatter(data['SL'],data['SW'],linewidths=0.5,color='red')
plt.title('Scatter chart')


plt.suptitle("Multiple Charts");
```

# **Box plot**

```python
plt.boxplot(data['SW'])
plt.title('Box Plot');
```

# **Line chart**

```python
bar_width = (max(data['SL']) - min(data['SL'])) / len(data['SL'])

plt.bar(data['SL'],data['SW'],width=bar_width,color='green')
plt.xlabel('SL')
plt.ylabel('SW')
plt.grid(True)
plt.title('Line chart ');
```

```python
plt.hist(data['SL'], bins=9, edgecolor='black')
plt.xlabel('SL')
plt.ylabel('Frequency')
plt.title('Distribution of SL')
plt.show()
```
